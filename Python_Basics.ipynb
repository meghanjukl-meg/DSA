#### Panlindrome

```python
string= "malayalam"
reverse = string[::-1]
if string == reverse:
    print("Panlindrome")
else:
    print("String is not panlindrome")
```

#### checking week

```python
day = 2
if day%7 ==0 :
    print("Sunday")
elif day%7 == 1 :
    print("Monday")
elif day%7 ==2 :
    print("Tuesday")
elif day%7 == 3:
    print("Wednesday")
elif day%7 == 4 :
    print("Thursday")
elif day%7 ==5 :
    print("Friday")
else:
    print("Saturday")
```

#### Convert to railway timing

```python
hour= 13
minute = 10
if hour >12 :
    hour = hour - 12
    print(f"{hour}:{minute} PM")
elif hour == 12:
    print(f"{hour}:{minute} PM")
else:
    print(f"{hour}:{minute} AM")
```
